```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# import and display data
df = pd.read_csv(r'C:\Users\Erin.Crump\Downloads\companies.csv')
df.head()
```

```python
# Question 1
# Using matplotlib/seaborn, create a countplot that shows a count of how many of the companies in the data set are from each state. 
# Paste a picture of the graph below. Note that you may need to adjust the width/height of the graph appropriately.
sns.countplot(x='state', data=df)
plt.xticks(rotation = 90)
plt.title('Count of Companies by State')
plt.show()
```

```python
# Question 3:
# Using matplotlib/seaborn, create a bar chart that shows a sum of how many total job openings there are in each industry.
sns.barplot(y='Industry', x='job_openings', data=df, errorbar=None, estimator=sum)
plt.title('Total Job Openings by Industry')
plt.show()
```

```python
# Question 5: Using matplotlib/seaborn, create a line chart that shows the average growth rate for each founded year.
sns.lineplot(x='founded', y='growth_percentage', data=df)
plt.xlabel('Year Founded')
plt.ylabel('Growth Rate')
plt.title('Average Growth per Year Founded')
plt.show()
```

```python
# Question 7: Using matplotlib/seaborn, create a scatter plot that compares the number of employees to estimated revenues for each company.
sns.scatterplot(y='estimated_revenues', x='employees', data=df, color='teal')
plt.show()
```

```python
# Question 9 Using matplotlib/seaborn, create a histogram that measures the number of companies founded in each year.
sns.histplot(x='founded', data=df, bins=20)
plt.title('Count of Years Founded')
plt.xlabel('Years Founded')
plt.show()
```
