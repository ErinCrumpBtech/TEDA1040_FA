```python
# Load in Data
import pandas as pd

df = pd.read_csv('invoices.csv')
```

```python
# Question 1: How many rows does the data set have?
df.shape[0]
```

```python
# Question 2: Which of the following columns contain null values?
# df.info()
df.columns[df.isnull().any()]
```

```python
# Question 3: Compute the descriptive statistics on the data set. Which columns appear to have logical errors?
df.describe()
```

```python
# Question 5: Fix standardization issues in the data set to determine which customer has created the most unique invoices. Which customer created the most invoices?
df['Customer'].value_counts()

df['Customer'] = df['Customer'].replace('Terry\'s Trinket', 'Terry\'s Trinkets')
df['Customer'] = df['Customer'].replace('Stuf by Stew', 'Stuff by Stew')
df['Customer'] = df['Customer'].replace('stuff by stew', 'Stuff by Stew')

df['Customer'].value_counts()
```

```python
# Question 6: Observe the null values in the LastEditedWhen column. Based on the number of null values in the column and the questions being asked by executives, what do you recommend doing to eliminate null values?
df['LastEditedWhen'].isnull().sum()
```

```python
# Question 7: Perform the action that you recommended above.
df.drop('LastEditedWhen', axis=1, inplace=True)

# After performing the action, which column contains the most null values?
df.isnull().sum()
```

```python
# Question 8: Observe the columns StockItemID and Description. Notice that StockItemID is a single number, and each number has a corresponding Description. Thus, by knowing the StockItemID of a row,
# its Description can be imputed. However, if both the StockItemID and Description are missing, the item is unknown.

# How many rows have both a null StockItemID and a null Description?
df[df['StockItemID'].isnull() & df['Description'].isnull()].shape[0]
```

```python
# Question 9: Answer: drop the rows with null values in the Description and StockItemID columns
df[['StockItemID', 'Description']].info()
```

```python
# Question 10: Perform the action that you recommended above. How many rows are left in the data frame?
df = df[~(df['StockItemID'].isnull() & df['Description'].isnull())]
# df.dropna(subset=['StockItemID', 'Description'], how='all')
df.shape[0]
```

```python
# Question 11: Answer: turn the negative tax rates into positive tax rates by multiplying by -1
df['TaxRate'].describe()
```

```python
# Question 12: Perform the action that you recommended above. What is the average TaxRate?
df['TaxRate'] = df['TaxRate'].abs()
round(df['TaxRate'].mean(),2)
```

```python
# Question 13: Answer: Impute the missing values
df['TaxAmount'].info()
```

```python
# Question 14: Perform the action that you recommended above. What is the average TaxAmount after removing the null values?
df['TaxAmount'].fillna((df['UnitPrice'] * df['Quantity'] * df['TaxRate']) / 100, inplace=True)
df['TaxAmount'].mean()
```

```python
# Question 15: Based on this knowledge, the number of null values in the data set, and the questions being asked by the executives, what do you recommend doing to remove null values from this column?
# Answer: Impute missing values
df['ExtendedPrice'].info()
```

```python
# Question 16: Perform the action that you recommended above. Treat the values of -999 as an input error and perform the recommended action on that value as well.
df['ExtendedPrice'].fillna(((df['UnitPrice'] * df['Quantity']) + df['TaxAmount']) , inplace=True)
df[(df['ExtendedPrice'] == -999),'ExtendedPrice'] = ((df['UnitPrice'] * df['Quantity']) + df['TaxAmount'])

# After addressing these issues, what is the average ExtendedPrice?
df['ExtendedPrice'].mean()
```

```python
# Question 17: Observe values in the Quantity and ExtendedPrice columns. Use the z-score method with a threshold of 3 to determine which rows are outliers in these columns.
df['Z-Score_Quantity'] = (df['Quantity'] - df['Quantity'].mean()) / df['Quantity'].std()
df['Z-Score_ExtendedPrice'] = (df['ExtendedPrice'] - df['ExtendedPrice'].mean()) / df['ExtendedPrice'].std()

# Based on the Quantity column, how many rows could be considered outliers?
df[(df['Z-Score_Quantity'] >= 3)].shape[0] # Note if you run this mutliple times after you already ran the code for Question 21 you will get 5922 because duplicate rows were dropped.
```

```python
# Question 18: Observe values in the Quantity and ExtendedPrice columns. Use the z-score method with a threshold of 3 to determine which rows are outliers in these columns.
# Make sure that you get rows that have a z-score greater than or equal to 3.
# Based on the ExtendedPrice column, how many rows could be considered outliers?
df[(df['Z-Score_ExtendedPrice'] >= 3)].shape[0]
```

```python
# Question 20: Observe the InvoiceLineID and the InvoiceID columns. The data set provided should have a unique ID for each invoice line, but there can be repeated invoice IDs because several different lines can be from the same invoice.
# Based on this knowledge, how would you handle checking for duplicate rows in the data set?
df[['InvoiceLineID','InvoiceID']].head(20)
```

```python
# Question 21: Perform the action that you recommended above and drop duplicate rows from the data set. How many total rows are in the data set after performing this action?
df.drop_duplicates(subset=['InvoiceLineID'], inplace=True)

df.shape[0]
```

```python
# Question 22: At this point, you should have a data set that is mostly free of null values (only has some null values in the Description column). Use this data set to answer the questions below.
df.isna().sum()

# Which item has the highest total quantity sold? Use the Description and Quantity columns.
df.groupby('Description').agg({'Quantity': 'sum'}).reset_index().sort_values(by='Quantity', ascending=False).head(1)
```

```python
# Question 23: Which item has generated the most total profit? Use the Description and LineProfit columns.
df.groupby('Description').agg({'LineProfit': 'sum'}).reset_index().sort_values(by='LineProfit', ascending=False).head(1)
```

```python
# Question 24: Which item is purchased in greatest quantities in a single order, on average? Use the Quantity and Description columns.
df.groupby('Description').agg({'Quantity': 'mean'}).reset_index().sort_values(by='Quantity', ascending=False).head(1)
```
